###What dimension should the vectors in our encoding be?

Everything in this notebook is assuming we are using the encoding that uses the second set of vectors, the ID-vectors, to get better performance, and that the cleanup memory is an associative cleanup memory.

First lets look at the distributions of vectors extracted from semantic pointers (using circular convolution and involution) as the number of dimensions change and the number of vectors in the trace change. This code iterates over Dimensions and Number of vectors, and for each draws samples from the distribution created by creating a semantic pointer trace with the given number of dimensions and given number of stored vectors, extracting one of the stored vectors from it, and comparing the extracted vector with the correct vector. The higher this number is, the better cleanup will work.

Also shown is a sampling of the dot-products of vectors drawn from the unit hypersphere.

```python
import numpy as np

# Number of dimensions
D = 2**np.arange(3,10)

# Number of vectors included in trace
N = [1, 5, 10]

num_samples = 100 # num samples for each histogram

vocab_size = 100
```

```python
#def make_term(vocab, vocabsize, termsize=2):
import random
from nengo.spa.pointer import SemanticPointer
from nengo.spa import Vocabulary
from nengo.utils.distributions import UniformHypersphere

data = []

for d in D:
    sub_data = []
    
    for n in N:
        sub_sub_data = []
        
        vocab = Vocabulary(d)
        for s in range(num_samples):
            indices = random.sample(range(vocab_size), 2 * n)
            
            sp = SemanticPointer(d)
            for i in range(n):
                sp += vocab.parse('X_'+str(indices[2*i])) * vocab.parse('X_'+str(indices[2*i+1]))
                
            sp.normalize()
            result = sp * ~vocab.parse('X_'+str(indices[0]))
            sub_sub_data.append(np.dot(result.v, vocab.parse('X_'+str(indices[1])).v))
        
        sub_data.append(np.array(sub_sub_data))
        
    data.append(sub_data)
    
randoms = [np.dot(UniformHypersphere(d, surface=True).sample(num_samples), 
                  UniformHypersphere(d, surface=True).sample(1).T) 
           for d in D]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(len(D), 1)

fig = plt.figure(figsize=(16,12))
i = 0
for d, sub_data, r in zip(D, data, randoms):
    plt.subplot(gs[i, 0])
    for n, sub_sub_data in zip(N, sub_data):
        plt.hist(sub_sub_data, bins=20, alpha=0.5, label="N = %d" % n)
        
    plt.hist(r, bins=20, alpha=0.5, label="Random")
    plt.gca().set_xlim((-0.4, 1.2))
    
    plt.legend(loc=1)
    plt.title("D = %d" % d)
    i += 1
plt.tight_layout()
```

We can see that the variance goes down as the dimension increases. With D = 512, we can reliably cleanup vectors extracted from traces with 5 stored pairs. We can see this by noting that almost all the samples are greater than 0.3, which is our chosen threshold. With lower dimensions (still with 5 pairs in the semantic pointer), there are a significant number of samples below 0.3, which will not get cleaned up.
