###What dimension should the vectors in our encoding be?

First lets look at the distributions of vectors extracted from semantic pointers (using circular convolution and involution) as we change the number of dimensions and number of bound pairs stored in the trace. This code iterates over dimensions and number of vectors, and for each draws samples from the distribution created by creating a semantic pointer trace with the given number of dimensions and given number of stored vectors, extracting one of the stored vectors from it, and comparing the extracted vector with the correct vector. As a rule of thumb, higher this number is, the better cleanup will work.

For example, for D = 512 and N = 3, for each sample the code will generate the semantic pointer $\mathbf{SP = x_1 \circledast x_2 + x_3 \circledast x_4 + x_5 \circledast x_6}$, where the $\mathbf{x_i}$ vectors are random 512-dimensional vectors on the unit hypersphere and are different for each sample. There are 3 bound pairs in $\mathbf{SP}$ because N = 3. It will then generate a data point by computing $\text{dot}(\mathbf{SP \circledast x_2^{-1}, x_1})$.

The fact that the $\mathbf{x_i}$ vectors are random unit vectors, rather than having structure of their own, means that we are mimicking the situation where we encode a semantic network by assigning each concept both a semantic pointer and an ID-vector, and the semantic pointers are defined in terms of ID-vectors of other concepts.

The reason we are interested in $\mathbf{SP \circledast x_2^{-1}}$ is because that is what will be input for the associative cleanup memory. In general, cleanup is only possible if $\text{dot}(\mathbf{SP \circledast x_2^{-1}, x_1})$ is greater than the threshold for the associative cleanup memory (usually 0.3).

```python
import numpy as np

# Number of dimensions
D = 2**np.arange(3,10)

# Number of vectors included in trace
N = [1, 3, 5]

num_samples = 1000 # num samples for each histogram

vocab_size = 100
```

```python
import random
import numpy as np
import warnings

from nengo.spa.pointer import SemanticPointer
from nengo.spa import Vocabulary
from nengo.utils.distributions import UniformHypersphere

data = []

# Ignore warnings about generating semantic pointers
warnings.simplefilter('ignore')

for d in D:
    sub_data = []
    
    for n in N:
        sub_sub_data = []
        
        vocab = Vocabulary(d)
        for s in range(num_samples):
            indices = random.sample(range(vocab_size), 2 * n)
            
            # Build up the semantic pointer
            sp = SemanticPointer(np.zeros(d))
            for i in range(n):
                sp += vocab.parse('X_'+str(indices[2*i])) * vocab.parse('X_'+str(indices[2*i+1]))
                
            sp.normalize()
            
            # Extract from semantic pointer
            result = sp * ~vocab.parse('X_'+str(indices[0]))
            
            # Compare extracted vector to correct vector
            sub_sub_data.append(np.dot(result.v, vocab.parse('X_'+str(indices[1])).v))
        
        sub_data.append(np.array(sub_sub_data))
        
    data.append(sub_data)
    
warnings.resetwarnings()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(len(D), 1)

fig = plt.figure(figsize=(16,12))
i = 0
for d, sub_data in zip(D, data):
    plt.subplot(gs[i, 0])
    for n, sub_sub_data in zip(N, sub_data):
        plt.hist(sub_sub_data, bins=20, alpha=0.5, label="N = %d" % n)
    
    plt.axvline(x=0.3, ls='--', c='k')
    plt.gca().set_xlim((-0.4, 1.2))
    
    plt.legend(loc=1)
    plt.title("D = %d" % d)
    i += 1
plt.tight_layout()
```

We can see that the variance goes down as the dimension increases. With D = 512, we can reliably cleanup vectors extracted from traces with 5 stored pairs. We can see this by noting that almost all the samples are greater than 0.3, which is our chosen threshold. For D below 256, a significant portion of the vectors extracted from traces with 5 bound pairs have similarity below 0.3 with the correct vector, meaning cleanup will fail (i.e., none of the populations in the associative cleanup memory will become active and the output of the cleanup memory will be 0).
